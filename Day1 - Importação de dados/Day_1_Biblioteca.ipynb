#Importando dados exemplares

```python
exemplares = pd.read_parquet('https://github.com/Lalset/Alura-7DaysToCode/blob/main/Day1%20-%20Importa%C3%A7%C3%A3o%20de%20dados/emprestimos/dados_exemplares.parquet?raw=true')
exemplares
```

#Unindo os dataframes

```python
emprestimos_unidos = emprestimos.merge(exemplares)
emprestimos_unidos
```

#Quantidade de Empréstimos por Ano
Houve um decaimento na quantidade de empréstimos por ano.

```python
emprestimos_unidos['ano'] = pd.to_datetime(emprestimos_unidos['data_emprestimo']).dt.year
```

```python
emprestimos_por_ano = emprestimos_unidos.groupby('ano').size()

emprestimos_por_ano.plot(kind='line', marker='o', title='Quantidade de Empréstimos por Ano')

plt.ylabel('Quantidade de Empréstimos')
plt.xlabel('Ano')
plt.grid(True)

plt.show()
```
